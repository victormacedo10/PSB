```python
# change to "%matplotlib notebook" to get interative plot with zoom functionalities 
%matplotlib inline
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mat4py import loadmat
from numpy import convolve as conv
from scipy.signal import convolve2d as conv2
from scipy.signal import *
from scipy.signal import convolve
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from pywfilters import wfilters
from IPython.display import display, HTML
from IPython.display import display, Math
np.set_printoptions(formatter={'float_kind':'{0:.4f}'.format})
CSS = """
.output {
    align-items: center;
}
"""

HTML('<style>{}</style>'.format(CSS))
```

```python
def downsample(x, factor):
    x = x[:len(x):factor]
    return x
def upsample(x, factor):
    y = np.zeros(len(x)*factor)
    y[:len(y):factor] = x
    return y
def down2d(x, factor=2):
    return x[:x.shape[0]:factor, :x.shape[1]:factor]
def up2d(x, factor=2):
    factor = int(factor)
    y = np.zeros([x.shape[0]*factor, x.shape[1]*factor])
    y[:y.shape[0]:factor, :y.shape[1]:factor] = x
    return y
def image_trim(x, rows, columns):
    rows_remove = x.shape[0] - rows;
    columns_remove = x.shape[1] - columns;
    half_rows_remove = rows_remove // 2
    half_columns_remove = columns_remove // 2
    y = x[half_rows_remove-1 : half_rows_remove + rows,  half_columns_remove-1 : half_columns_remove+columns]
    return y
def iterate2dfilters(h, downsample_factors, h_ll, compensatory_delays= None, d=1):
    if compensatory_delays == None:
        compensatory_delays = np.array([0, 0, 0, 0])
    hi = [0]*(len(h) + 3)
    hi[len(h) + 3 - 1] = h[len(h)-1]
#     % n = compensatory_delays(end) / d + 1;
    downsample_factors_i = np.append(np.zeros(3), downsample_factors)
    compensatory_delays_i = np.append(compensatory_delays, np.zeros(3) + 2 * compensatory_delays[-1] + d)
    for n in range(5-1, len(h) + 2):
        hi[n] = h[n-3]
        
    hi[0] = conv2(h_ll, up2d(np.array(h[0])))
    hi[1] = conv2(h_ll, up2d(np.array(h[1])))
    hi[2] = conv2(h_ll, up2d(np.array(h[2])))
    hi[3] = conv2(h_ll, up2d(np.array(h[3])))
    downsample_factors_i[:4] = downsample_factors[0] * 2
    return hi, downsample_factors_i, compensatory_delays_i

def qmf_decomposition_2d(x, h0, h1, levels):
    h_ll, h_hl, h_lh, h_hh = np.outer(h0, h0), np.outer(h1, h0), np.outer(h0, h1), np.outer(h1, h1)
    h = [0]*4
    h[0], h[1], h[2], h[3] = h_ll, h_hl, h_lh, h_hh
    downsample_factors = np.array([2,2,2,2])
    for n in range(1, levels):
        h, downsample_factors, _ = iterate2dfilters(h, downsample_factors, h_ll)
    n = len(h) - 1
    xdc = [0]*(n+1)  
    xdc[n] = down2d(convolve(np.array(h[n]), x, mode='full'), int(downsample_factors[n]))
    for n in range(len(h) - 1):
        xdc[n] = down2d(convolve(np.array(h[n]), x, mode='full'), int(downsample_factors[n]))
    xd = np.zeros(x.shape)
    n = len(xdc) - 1
    central_row = x.shape[0] - 1
    central_column = x.shape[1] -1 
    while n > 0:
        central_row = central_row // 2
        central_column = central_column // 2
        xdcn = image_trim(xdc[n], central_row, central_column)
        xdcn1 = image_trim(xdc[n - 1], central_row, central_column)
        xdcn2 = image_trim(xdc[n - 2], central_row, central_column)
        xd[central_row:2*central_row+1, central_column:2*central_column+1] = np.abs(xdcn) / np.max(np.abs(xdcn))
        xd[: central_row+1, central_column: 2*central_column+1] = np.abs(xdcn2) / np.max(np.abs(xdcn2))
        xd[central_row : 2*central_row+1, : central_column+1] = np.abs(xdcn1) / np.max(np.abs(xdcn1))
        n = n - 3

    xdcn = image_trim(xdc[0], central_row, central_column)
    xd[: central_row+1, : central_column+1] = np.abs(xdcn) / np.max(np.abs(xdcn))
    return xd, xdc, h

def poly_sum(x, y):
    x1 = np.zeros(max(len(x), len(y)))
    y1 = np.copy(x1)
    x1[:len(x)] = x
    y1[:len(y)] = y
    z = x1 + y1
    return z

def evaluate_decomposition_synthesis_filters(h0, h1, g0, g1, tol=1e-6):
    h0_ = np.copy(h0)
    h0_[1:len(h0_):2] = -h0_[1:len(h0_):2]
    h1_ = np.array(h1)
    h1_[1:len(h1_):2] = -h1_[1:len(h1_):2]
    alias_term = poly_sum(conv(h0_, g0), conv(h1_, g1))
    alias_term_tol = np.where(np.abs(alias_term) < tol, 0, alias_term)
    cond1 = np.sum(alias_term_tol) == 0
    if cond1:
        print("Primeira condição satisfeita!")
    else:
        print("Primeira condição não satisfeita")
    h0g0 = conv(h0, g0)
    h1g1 = conv(h1, g1)
    lti_term = poly_sum(h0g0, h1g1)/2
    lti_term_tol = np.where(np.abs(lti_term) < tol, 0, lti_term)
    k = np.where(lti_term_tol != 0)
    cond2 = len(k) == 1
    if cond2:
        print("Segunda condição satisfeita!")
    else:
        print("Segunda condição não satisfeita")
        
    perfect_reconstruction = cond2 and cond1

    A = int(round(lti_term_tol[int(k[0][0])]))
    d = int(k[0][0])
    print(d)
    return perfect_reconstruction, A, d

def image_delay(x, rows, columns):
    y = np.zeros([x.shape[0] + int(rows), x.shape[1] + int(columns)])
    y[int(rows) : y.shape[0], int(columns) : y.shape[1]] = x
    return y

def add_images(x, y):
    rows = max([x.shape[0], y.shape[0]])
    columns = max([x.shape[1], y.shape[1]])
    x_ = np.zeros([rows, columns])
    x_[:x.shape[0], :x.shape[1]] = x
    y_ = np.zeros([rows, columns])
    y_[:y.shape[0], :y.shape[1]] = y
    z = x_ + y_
    z = z[:rows, :columns]
    return z

def qmf_reconstruction_2d(xdc, h0, h1, g0, g1, rows, columns):
    _, A, d = evaluate_decomposition_synthesis_filters(h0, h1, g0, g1)
    levels = int((len(xdc) - 1) / 3)
    g_ll, g_hl, g_lh, g_hh = np.outer(g0, g0), np.outer(g1, g0), np.outer(g0, g1), np.outer(g1, g1)
    g = [0]*4
    g[0], g[1], g[2], g[3] = g_ll, g_hl, g_lh, g_hh
    upsample_factors = np.array([2,2,2,2])
    compensatory_delays = np.array([0,0,0,0])
    for n in range(1, levels):
        g, upsample_factors, compensatory_delays = iterate2dfilters(h, upsample_factors, h_ll)
    
    xr = np.zeros([1,1])
    y = [0]*len(xdc)
    for n in range(len(xdc)-1):
        # print(n)
        x = xdc[n]
        x = image_delay(x, compensatory_delays[n], compensatory_delays[n])
        x = up2d(x, upsample_factors[n])
        x = convolve(np.array(g[n]), x, mode='full')
        xr = add_images(xr, x)
        y[n] = x
    y = np.array(y)
    xdelay = xr
    compensatory_delay = 2 * compensatory_delays[-1] + d
    xr = xr[compensatory_delay:, compensatory_delay:]
    xr = xr[:rows, :columns]

    return xr, xdelay, compensatory_delay, y


def plot_filters_reponses(h0, h1, g0, g1):
    h_names = ['h0', 'h1', 'g0', 'g1']
    h = [h0, h1, g0, g1]
    n_filters = len(h)
    fig = plt.figure(figsize=[13,4])
    ax1 = fig.add_subplot(121)
    for i in range(n_filters):
        w, h_mag = freqz(h[i])
        plt.plot(w, abs(h_mag), label = h_names[i])
    plt.ylabel('Magnitude')
    plt.xlabel('Frequencies [rad/sample]')
    plt.grid()
    plt.legend()

    ax2 = fig.add_subplot(122)
    for i in range(n_filters):
        w, h_mag = freqz(h[i])
        angles = np.unwrap(np.angle(h_mag))
        plt.plot(w, angles, label = h_names[i])
    plt.ylabel('Angle (radians)')
    plt.xlabel('Frequencies [rad/sample]')
    plt.grid()
    plt.legend()
    plt.show()
    
def check_ortogonality(h0, h1, g0, g1):
    if(np.sum(np.dot(h0, h1)) == 0):
        print("Os filtros de decomposição são ortogonais")
    if(np.sum(np.dot(g0, g1)) == 0):
        print("Os filtros de reconstrução são ortogonais")
```
