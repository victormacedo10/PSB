This notebook presents the code to analyze an electrocardiogram signal (ECG) and extract useful features. With this goal, a sample ECG signal is first plotted in order to get a "feeling" of how the signal behaves and what patterns can be exploited. 

Then, a simple and limited algorithm is implemented to identify and separate the QRS complex. All the segmented QRS complex signals are utilized to form two different signals: 

1. A signal composed of all the QRS complex segments concatenated.
2. A similiar signal to the original but with zeros in the regions outside the identified QRS complex segments.
    
Finally, to check how the power of the signal is distributed in the frequency spectrum, the PSD (Power Spectrum Density) is calculated and plotted. The PSD will be analyzed for original signal, the two generated signals (mentioned above) and for a single windowed QRS complex segment.

In order to organize the notebook and make the data analysis clearer, the develop functions are placed in a separated section from the visualizations and discussions.

# Functions

All the implemented functions used are presented in this section.

```python
def plotECGSignal(ecg_signal, fs, t0=0, delta_t=-1):
    """
    This function plots a signal for a given time interval.

    If no interval if given the whole signal will be plotted.

    If the interval exceeds the signal duration, a warning will be printed in the terminal

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    fs : integer
        The signal sampling rate in Hertz
    t0 : float
        The time to begin plotting in seconds
    delta_t : float
        The duration of the signal to be plotted in seconds
    """
    if delta_t == -1: # if delta is -1 plot till the end of the signal
        delta_t = len(ecg_signal)*(1/fs) - t0 # set delta to the duration of the signal minus t0
    
    duration = len(ecg_signal)*(1/fs) # calculate duration from num of samples and sampling rate
    calc_duration = t0 + delta_t # get input required duration
    if calc_duration > duration: # check if required duration exceeds the real duration
        print("Exceeded signal duration ({} s) in {} s, lower t0 or delta_t".format(duration, calc_duration - duration))
        return
    n0 = int(round(t0*fs)) # get equivalent sample for the initial time
    n1 = n0 + int(round(delta_t*fs)) # get equivalent sample for the final time
    # produce time vector to show plot in sec
    t = np.linspace(0, len(ecg_signal)/fs, len(ecg_signal))
    # configure plot for the ECG signal
    plt.figure()
    plt.plot(t[n0:n1], ecg_signal[n0:n1])
    plt.ylabel("ECG amplitude")
    plt.xlabel("time (s)")
    plt.grid()
    plt.show()
```

```python
def plotECGPeaks(ecg_signal, Q_cplx, R_cplx, S_cplx, fs, t0=0, delta_t=-1):
    """
    This function plots the ECG highlighting its peaks for a given time interval.

    If no interval if given the whole signal will be plotted.

    If the interval exceeds the signal duration, a warning will be printed in the terminal

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    Q_cplx : numpy array (2,)
        A vector with the Q peak value and position
    R_cplx : numpy array (2,)
        A vector with the R peak value and position
    S_cplx : numpy array (2,)
        A vector with the S peak value and position
    fs : integer
        The signal sampling rate in Hertz
    t0 : float
        The time to begin plotting in seconds
    delta_t : float
        The duration of the signal to be plotted in seconds
    """
    if delta_t == -1: # if delta is -1 plot till the end of the signal
        delta_t = len(ecg_signal)*(1/fs) - t0 # set delta to the duration of the signal minus t0
    
    duration = len(ecg_signal)*(1/fs) # calculate duration from num of samples and sampling rate
    calc_duration = t0 + delta_t # get input required duration
    if calc_duration > duration: # check if required duration exceeds the real duration
        print("Exceeded signal duration ({} s) in {} s, lower t0 or delta_t".format(duration, calc_duration - duration))
        return
    n0 = int(round(t0*fs)) # get equivalent sample for the initial time
    n1 = n0 + int(round(delta_t*fs)) # get equivalent sample for the final time
    n = np.arange(n0, n1)
    # configure plot for the ECG signal and peaks
    plt.figure()
    plt.plot(n, ecg_signal[n0:n1])
    plt.scatter(R_cplx[:,1], R_cplx[:,0], marker='o') # plot positions for the R wave peak
    plt.scatter(S_cplx[:,1], S_cplx[:,0], marker='o') # plot positions for the S wave peak
    plt.scatter(Q_cplx[:,1], Q_cplx[:,0], marker='o') # plot positions for the R wave peak
    plt.xlim([n0, n1]) # set limit of the plot as defined by t0 and delta_t
    plt.ylabel("ECG amplitude")
    plt.xlabel("samples (n)")
    plt.grid()
    plt.show()
```

```python
def plotAllQRSSignals(signal_vec, fs):
    """
    This function plots all signals in a vector.

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    fs : integer
        The signal sampling rate in Hertz
    """
    # configure all QRS plot
    plt.figure()
    for i in range(len(signal_vec)): # iterate through each QRS segment
        # produce time vector to show plot in sec
        t = np.linspace(0, len(signal_vec[i])/fs, len(signal_vec[i]))
        plt.plot(t, signal_vec[i]) # plot individual signal
    plt.ylabel("ECG amplitude")
    plt.xlabel("time (s)")
    plt.grid()
    plt.show()
```

```python
def getSfromR(ecg_signal, R, S_thres=0):
    """
    This function finds the S peak given a detected R peak and a threshold that distincts the S wave

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    R : numpy array (2,)
        A vector with the R peak value and position
    S_thres : float
        The threshold to distinct the S wave
        
    Returns
    -------
    S : numpy array (2,)
        A vector with the S peak value and position
    """
    i = R[1] # set i to the index of the R peak
    # iterate forward in the signal till value crosses the defined threshold
    while(ecg_signal[i] > S_thres):
        i+=1
    # local min starts in the first value lower than the threshold
    local_min = [float(ecg_signal[i]), i]
    # iterate forward in the signal while value remains in the defined threshold             
    while(ecg_signal[i] < S_thres): 
        if(ecg_signal[i] < local_min[0]): # check if value if lower than the current minimum
            local_min = [float(ecg_signal[i]), i] # set value as new local minimum
        i+=1 # iterate to get next sample
    S = local_min # S peak is the lowest peak detected
    return S
```

```python
def getQfromR(ecg_signal, R, Q_thres=0):
    """
    This function finds the Q peak given a detected R peak and a threshold that distincts the Q wave

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    R : numpy array (2,)
        A vector with the R peak value and position
    Q_thres : float
        The threshold to distinct the Q wave
        
    Returns
    -------
    Q : numpy array (2,)
        A vector with the S peak value and position
    """
    i = R[1] # set i to the index of the R peak
    # iterate backwards in the signal till value crosses the defined threshold
    while(ecg_signal[i] > Q_thres):
        i-=1
    # local min starts in the first value lower than the threshold
    local_min = [float(ecg_signal[i]), i]
    # iterate backwards in the signal while value remains in the defined threshold 
    while(ecg_signal[i] < Q_thres):
        if(ecg_signal[i] < local_min[0]): # check if value if lower than the current minimum
            local_min = [float(ecg_signal[i]), i] # set value as new local minimum
        i-=1 # iterate to get next sample
    Q = local_min # Q peak is the lowest peak detected
    return Q
```

```python
def getQRSfromQandS(ecg_signal, Q_idx, S_idx, QRS_thres=0):
    """
    This function returns the limits of the QRS complex given the Q and S peak indexes and a threshold to crop the complex

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    Q_idx : integer
        The index of the respective Q wave
    S_idx : integer
        The index of the respective S wave
    QRS_thres : float
        The threshold to distinct the Q wave
        
    Returns
    -------
    [i_min, i_max] : list (2)
        A list with the minimum and maximum limit of the QRS complex in question
    """
    i_min = Q_idx # set starting minimum index for the interval as the Q peak index
    i_max = S_idx # set starting maximum index for the interval as the S peak index
    # iterate backwards to find the first value to exceeds the given threshold
    while(ecg_signal[i_min] < QRS_thres):
        i_min-=1
    # iterate fowards to find the first value to exceeds the given threshold
    while(ecg_signal[i_max] < QRS_thres):
        i_max+=1
    return [i_min, i_max]
```

```python
def getQRSsegments(ecg_signal, R_thres=0.3, Q_thres=0, S_thres=0, QRS_thres=0):
    """
    This function finds the R, Q and S peaks, given the respective thresholds, and returns all the detected peaks, the QRS complex concatenated and the QRS complex with zeros

    Parameters
    ----------
    ecg_signal : numpy array (signal_length,)
        The ecg signal as a row vector
    R_thres : float
        The threshold to distinct the R wave
    Q_thres : float
        The threshold to distinct the Q wave
    S_thres : float
        The threshold to distinct the S wave
    QRS_thres : float
        The threshold to segment the QRS complex
        
    Returns
    -------
    QRS_cplx_zeros : numpy array (signal_length,)
        A vector with the QRS complex with zeros
    QRS_cplx_conc : numpy array (all QRS complex lengths,)
        A vector with the QRS complex concatenated
    QRS_cplx : list (num of QRS complex, QRS_complex lenght)
        A list with the QRS complex append as numpy arrays
    Q_cplx : numpy array (2,)
        A vector with the Q peak value and position
    R_cplx : numpy array (2,)
        A vector with the R peak value and position
    S_cplx : numpy array (2,)
        A vector with the S peak value and position
    """
    # signal with zeros to store the QRS segments completed with zeros
    QRS_cplx_zeros = np.zeros(ecg_signal.shape)
    QRS_cplx = [] # list to store each QRS segment identified
    conc_len = 0 # variable to store the lenght of the QRS segments concatenated

    Q_cplx = [] # list to store each Q peak value and index
    R_cplx = [] # list to store each R peak value and index
    S_cplx = [] # list to store each S peak value and index

    i=0 # start counter to iterate through the signal
    local_min = [float(ecg_signal[0]), 0] # start local minimum counter as the first sample
    while(i<len(ecg_signal)): # iterate through the whole signal to find R waves
        # if signal value exceeds R threshold start QRS segmentation process
        if(ecg_signal[i] > R_thres):
            # start local max as the first value higher than the R threshold
            local_max = [float(ecg_signal[i]), i]
            i+=1 # get next sample
            # if signal value exceeds R threshold find next local max
            while(ecg_signal[i] > R_thres):
                # if current sample is higher than the previous local max, update local max
                if(ecg_signal[i] > local_max[0]):
                    local_max = [float(ecg_signal[i]), i]
                i+=1 # get next sample
            # append to R peaks vector the local max
            R_cplx.append(local_max)
            # get the S peak from the R peak
            local_min_S = getSfromR(ecg_signal, local_max, S_thres)
            # append to S peaks vector the current detected S peak
            S_cplx.append([ecg_signal[local_min_S[1]],local_min_S[1]])
            # get the Q peak from the R peak
            local_min_Q = getQfromR(ecg_signal, local_max, Q_thres)
            # append to Q peaks vector the current detected Q peak
            Q_cplx.append([ecg_signal[local_min_Q[1]],local_min_Q[1]])
            # get the QRS interval from the detected Q and S peaks
            QRS_idx = getQRSfromQandS(ecg_signal, local_min_Q[1], local_min_S[1], QRS_thres=0)
            # append to QRS vector the current QRS segment
            QRS_cplx.append(ecg_signal[QRS_idx[0]:QRS_idx[1]])
            # insert the QRS segment in the QRS segments with zeros signal
            QRS_cplx_zeros[QRS_idx[0]:QRS_idx[1]] = ecg_signal[QRS_idx[0]:QRS_idx[1]]
            # update the QRS segments concatenated signal length
            conc_len += len(ecg_signal[QRS_idx[0]:QRS_idx[1]])
        i+=1 # get next sample after the last QRS segment detected
        
    # transform the each peaks list to a numpy array to allow further operations
    R_cplx = np.array(R_cplx)
    S_cplx = np.array(S_cplx)
    Q_cplx = np.array(Q_cplx)
    # signal with zeros to store the QRS segments concatenated
    QRS_cplx_conc = np.zeros(conc_len)
    pos = 0 # variable to store positions for the QRS segments concatenated
    j = 0 # varible to iterate through each QRS segment detected
    for j in range(len(QRS_cplx)): # iterate through each QRS segment detected
        # insert the QRS segment in the QRS segments concatenated signal
        QRS_cplx_conc[pos:pos + len(QRS_cplx[j])] = QRS_cplx[j]
        pos = pos + len(QRS_cplx[j]) # update the position based on the current segment length
    return QRS_cplx_zeros, QRS_cplx_conc, QRS_cplx, R_cplx, S_cplx, Q_cplx
```

```python
def plotPSDSignal(signal, fs):
    """
    This function plots a signal power espectrum density (PSD)

    Parameters
    ----------
    esignal : numpy array (signal_length,)
        The signal as a row vector
    fs : integer
        The signal sampling rate in Hertz
    """
    N = len(signal) # store the number of sample in the signal
    # create a frequency vector to match the frequency index for each sample
    f = np.linspace(0, fs, N)
    # apply the fft transform to the signal normalized by the signal length
    signal_fft = np.fft.fft(signal)/N
    # calculate the PSB by elevating the signal magnitude to the second power
    psd = np.power(np.abs(signal_fft), 2)
    # configure plot
    plt.figure() 
    plt.plot(f[:N // 2], psd[:N // 2]) # plot only the positive frequencies
    plt.xlabel("frequency (Hz)")
    plt.grid()
    plt.show()
```

```python
def plotAllPSDSignals(signal_vec, fs):
    """
    This function plots a signal power espectrum density (PSD)

    Parameters
    ----------
    esignal : numpy array (signal_length,)
        The signal as a row vector
    fs : integer
        The signal sampling rate in Hertz
    """
    # configure All PSD plot
    plt.figure() 
    for i in range(len(signal_vec)): # iterate through each given signal
        N = len(signal_vec[i]) # store the number of sample in the signal
        f = np.linspace(0, fs, N) # create a frequency vector to match the frequency index for each sample
        # apply the fft transform to the signal normalized by the signal length
        signal_fft = np.fft.fft(signal_vec[i])/N
        # calculate the PSB by elevating the signal magnitude to the second power
        psd = np.power(np.abs(signal_fft), 2)
        plt.plot(f[:N // 2], psd[:N // 2]) # plot only the positive frequencies
    plt.xlabel("frequency (Hz)")
    plt.grid()
    plt.show()
```

# ECG Analysis

In this section the ECG sample signal is analyzed to identify the Q, R and S locations and later segment the QRS complex.

## Raw ECG plot

The signal used for the experiment is the first sample named "ecg01.mat" which has a sampling frequency fs of 125 Hz.

```python
data_input = loadmat('sinais_ecg/ecg01.mat') # load signal
# get only the x values and transform in a row vector
data_ecg = np.array(data_input['x'])[:, 0]
fs = 125 # set sampling rate
```

In order to plot in a visible scale, only an interval of a couple ECG cycles are displayed. To control the region to be displayed the starting point t0 can be defined, along with lenght in seconds desired.

```python
plotECGSignal(data_ecg, fs, t0=10, delta_t=5)
```

## QRS Segmentation

By analysing the ECG example plotted, we can see that this particular signal tends to maintain its offset and doesn't drift over time. At the same time, all the waves seem to have similiar amplitudes. For these reason, a simple algorithm based on threshold will be developed.

First, we can find the most distinguished wave peak, the R wave. Scanning through the signal we can see that all the R peaks exceeds a 0.3 threshold. So, by iterating through the signal the algorithm can search for disconnected segments where its values are above the threshold and then consider the R peak to be the maximum in that interval.

With the R peak found, a similiar process can be done to find the Q and S peaks. Iterating from the detected R peak, we search backwards for the first segment below the Q threshold defined to be 0 and fowards for the first segment below the S threshold also defined to be 0, the Q and S peaks are then defined as the minimum of those segments respectively.

Finally, if the Q and S peaks found, another similiar process is implemented to found the QRS complex endings, where from the Q peak the complex will be cropped left in the first zero found backwars and cropped right in the first zero found fowards from the S peak.

The algorithm then returns all the Q, R and S peaks found (with their values and positions) in separate arrays and another array with all the segments separated in rows. Also, the two QRS signals (completed with zeros and appended) are formed and returned.

```python
QRS_cplx_zeros, QRS_cplx_conc, QRS_cplx, R_cplx, S_cplx, Q_cplx = getQRSsegments(data_ecg, R_thres=0.3, Q_thres=0, S_thres=0)
```

### QRS identified

Now we plot the detected peaks to see the result. As depicted in the graph below, all the peak were correctly detected for this region.

```python
plotECGPeaks(data_ecg, Q_cplx, R_cplx, S_cplx, fs, t0=10, delta_t=5)
```

### QRS separated

In order to see if the segmented QRS complex were similiar, we plotted all the waves in a single graph. As ilustrated below, only one wave had a late detection in the Q peak and the complex appeared slightly shifted to the left.

```python
plotAllQRSSignals(QRS_cplx, fs)
```

Then the concatened QRS complex signal is plotted to see if the result matches what is expected.

```python
plotECGSignal(QRS_cplx_conc, fs, t0=10, delta_t=5)
```

And, finally, the QRS complex signal completed with zeros is also plotted.

```python
plotECGSignal(QRS_cplx_zeros, fs, t0=10, delta_t=5)
```

The main flaw of the algorithm, asides the dependency in defining the thresholds, is that its not currently robust to inverted ECG signals, which are commonly used. But, this adaptation is not difficult and can be done (in some cases) by comparing the magnitude of the minimum and maximum values and deciding whether or not to invert the signal before passing through the QRS segmentation function and also remembering to invert back after returning the segmented arrays.

Overall , the results in this section suggests that the QRS segmentation and signals formation were done sucessfully for the main part. Though some mistakes were enconteured, they shouldn't compromise the data analysis, which is the next section.

# Power Analysis

In this section all the resulting signal's PSD will be plotted and analyzed.

## Original Signal

First we plot the original ECG signal PSD, for comparison. As ilustrated below we have most of the signal power below 10 Hz.

```python
plotPSDSignal(data_ecg, fs)
```

Similiarly, the QRS segments PSD's are also mainly below 10 Hz, but now, because de signal has less points and the FFT was calculated with same amount of points, the PSD appeared in a lower frequency resolution.

## Individual QRS segments

To check the power distribution over the frequencies for the individual QRS segments we plotted all their PSD simultaneously below. Similiar to the original the main part of the signal is located below 10 Hz, but in this case, because the FFT was calculated with the same number of points as the input signal, its frequency resolution was lower, making it harder to distinct between the lower frequencies.

```python
plotAllPSDSignals(QRS_cplx, fs)
```

## QRS Concatenated

Then, for the signal with all the QRS segments concatenated, the PSD showed a similiar power distribution as the original, but with less peaks and the density more spreaded out in the frequencies. This should be due to the high frequencies introduced by each segment window transition.

```python
plotPSDSignal(QRS_cplx_conc, fs)
```

To better visualize the differences between the original and concatenated signals, we plotted them together below. In this case, the graph shows that the mean peaks changed by eliminating the regions without a QRS segment.

```python
plotAllPSDSignals([data_ecg, QRS_cplx_conc], fs)
```

## QRS with Zeros

Finally, the signal where values were substituted with zeros in the regions outside the QRS segments showed a power distribution if closer to the original, but with more evenly distributing of power through the peaks in the lower frequencies, which consequently reduced the max peaks amplitude.

```python
plotPSDSignal(QRS_cplx_zeros, fs)
```

To check if the peaks location were mantained between the original signal and the signal with zero we plotted both simultaneously. As we can see, the peaks seem to match and only the amplitude changed, which makes sense since we kept most of the signal content intact.

```python
plotAllPSDSignals([data_ecg, QRS_cplx_zeros], fs)
```
